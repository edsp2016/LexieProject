```python
#import urllib2
import urllib.request as ul
import bs4
import sys
import time
import json
from bs4 import BeautifulSoup
```

```python
import pandas as pd
Location=r'imdb_rating.xls'
df = pd.read_excel(Location)
df.head()
```

```python
urls=[]
for i in range(len(df)):
    html="http://www.imdb.com/title/"+df.Imdb_id[i]
    #url=ul.urlopen(html).read()
    urls.append(html)
urls
    
```

```python
sum_dire=[]
for html in urls:
    #url=ul.urlopen(html).read()
    soup = bs4.BeautifulSoup(ul.urlopen(html).read())
    director=[]
    try:
        dir_table=soup.find_all("div", class_="credit_summary_item")[0]
        dir_list=dir_table.find_all("span", { "class" : "itemprop" })
        for i in range(len(dir_list)):
            director.append(dir_list[i].text)
    except: IndexError
        #director=['NA']
    sum_dire.append(director)
    
    

```

```python
director_1=[]
director_2=[]
for i in range(len(sum_dire)):
    director=sum_dire[i]
    #director1,director2=director.split(",")
    if len(director)>1:
        director1=director[0]
        director2=director[1]
    else:
        director1=director[0]
        #director2=""   
    director_1.append(director1)
    director_2.append(director2)
```

```python
def parse_table(table):
    """ Get data from table """
    return [
        [cell.get_text().strip() for cell in row.find_all(['th', 'td'])]
           for row in table.find_all('tr')
    ]
```

```python
star_1=[]
star_2=[]
for i in range(len(urls)):
    try:
        page = BeautifulSoup(ul.urlopen(urls[i]))
        table = page.find_all('table')
    #star=["",""]
        s1=parse_table(table[0])[1][1]
        s2=parse_table(table[0])[2][1]
    except: IndexError
        s1='NA'
        s2='NA'
    star_1.append(s1)
    star_2.append(s2)
    
        
    
    

```

```python
df["star1"]=star_1
df["star2"]=star_2
df["director1"]=director_1
df["director2"]=director_2
del df['directors']
df.head()
```
