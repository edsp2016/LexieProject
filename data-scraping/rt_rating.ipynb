```python
def parse_table(table):
    """ Get data from table """
    return [
        [cell.get_text().strip() for cell in row.find_all(['th', 'td'])]
           for row in table.find_all('tr')
    ]
```

```python
urls=[]
for i in range(1988,2016):
    url="http://www.rottentomatoes.com/top/bestofrt/?year="+str(i)
    urls.append(url)
#urls
```

```python
tomato=[['Rank', 'RatingTomatometer', 'Title', 'No. of Reviews'],]
```

```python
for i in range(1988,2016):
    url="http://www.rottentomatoes.com/top/bestofrt/?year="+str(i)
    soup = bs4.BeautifulSoup(ul.urlopen(url).read())
    table=soup.find_all('table')
    tomato.append(parse_table(table[2])[1:])
    
```

```python
tomato
```

```python
tomatoes=[['Rank', 'RatingTomatometer', 'Title', 'No. of Reviews'],]
for i in range(1,len(tomato)):
    for j in range(len(tomato[i])):
        tomatoes.append(tomato[i][j])
    
    
```

```python
tomatoes
```

```python
rank=[]
tomato_rating=[]
title=[]
reviewnum=[]
for i in range(1,len(tomatoes)):
    rank.append(tomatoes[i][0])
    tomato_rating.append(tomatoes[i][1])
    title.append(tomatoes[i][2])
    reviewnum.append(tomatoes[i][3])

    
```

```python
d={"rank":rank,"tomato_rating":tomato_rating,"title":title,"reviewnum":reviewnum}
df2 = pd.DataFrame(data=d)
```
