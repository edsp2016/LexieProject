```python
import urllib.request as ul
import bs4
import sys
import time
import json
from bs4 import BeautifulSoup
```

```python
import pandas as pd
Location=r'movies_people.xlsx'
df = pd.read_excel(Location)
df.head()
```

```python
from imdbpie import Imdb
imdb = Imdb()
imdb = Imdb(anonymize=True) # to proxy requests
imdb.search_for_person("Sam Worthington")[0]['imdb_id']
```

```python
#get directors and actors IMDb id
dir_id=[]
#star1_id=[]
#star2_id=[]
for i in range(len(df)):
    dir_id.append(imdb.search_for_person(df.director1[i])[0]['imdb_id'])
    #star1_id.append(imdb.search_for_person(df.star1[i])[0]['imdb_id'])
    #star2_id.append(imdb.search_for_person(df.star2[i])[0]['imdb_id'])
    
```

```python
df['dir_id']=dir_id
#df['star1_id']=star1_id
#df['star2_id']=star2_id
```

```python
#get first five reviews from each movie
rew=[]
for i in range(1,100):
    review=[]
    for m in range(1,5):
        review.append(imdb.get_title_reviews(df.Imdb_id[i], max_results=5)[m].text)
    rew.append(review)
    
#reviews=[]
#for m in range(1,5):
    #reviews.append(imdb.get_title_reviews("tt0468569", max_results=5)[m].text)
    
```

```python
star1_id=[]
star2_id=[]
for i in range(len(df)):
    try:
        star1_id.append(imdb.search_for_person(df.star1[i])[0]['imdb_id'])
        star2_id.append(imdb.search_for_person(df.star2[i])[0]['imdb_id'])
    except IndexError:
        star1_id.append("")
        star2_id.append("")   
```

```python
df['star1_id']=star1_id
df['star2_id']=star2_id
df.to_excel('IMDb0425.xlsx', index=False)
print('Done')
```
