Grab Data from Excel
We will be using the read_excel function to read in data from an Excel file. The function allows you to read in specfic tabs by name or location.

```python
from imdbpie import Imdb
import pandas as pd
```

```python
imdb = Imdb()
imdb = Imdb(anonymize=True) # to proxy requests
imdb = Imdb(cache=True)
```

```python
# Parse a specific sheet
Location=r'600movies.xlsx'
df = pd.read_excel(Location)
df.dtypes
```

```python
df.head()
```

Prepare Data

```python
#function to find imdb movie id from a movie name
def imdbid(movie_title):
    movie_id=imdb.search_for_title(movie_title)[0]['imdb_id']
    return movie_id
##function to get rating and type
#imdb_id = imdb.get_title_by_id(movie_id)
#imdb_id.rating
#imdb_id.type
```

```python
#imdb.search_for_title("Avatar")[0]['imdb_id']
```

```python
title=df['Title']
#title[0]
imdb_id_list=[]
imdb_id_rating_list=[]
#imdb_id_type_list=[]
for i in range(len(title)):
    movie=imdbid(title[i])
    imdb_id_list.append(movie)
    rating=imdb.get_title_by_id(movie).rating
    #movie_type=imdb.get_title_by_id(movie).type
    imdb_id_rating_list.append(rating)
    #imdb_id_type_list.append(movie_type)

    
```

```python
print(imdb_id_list)
imdb_id_rating_list
#imdb_id_type_list
#len(imdb_id_list)=633
```

```python
# Lets add a column
df['Rating'] = imdb_id_rating_list
df['Imdb_id']= imdb_id_list
df
```

```python
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
df['Rating'].plot(figsize=(15,5));

```

Export to EXCEL

```python
df.to_excel('imdb_rating.xls', index=False)
print('Done')
```
